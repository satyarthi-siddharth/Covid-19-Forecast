```
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Data Science/Data For Data_Science & Machine_Learning/Covid-19_Data.csv')
df
```

```
df.index = pd.PeriodIndex(df.date,freq='D')
df.index.rename('Date',inplace=True)
```

```
df_ind = df[df['location']=='India']
df_ind
```

```
df_ind.columns
```

```
df_ind.info()
```

```
df_ind.dropna(subset=['total_cases','total_deaths'],inplace=True)
```

```
df_ind
```

```
df_ind.info()
```

```
df_ind['date'] = pd.to_datetime(df_ind['date'])
df_ind
```

```
df_ind.info()
```

```
df_ind_new = df_ind[['date','new_cases','new_deaths']]
df_ind_new
```

```
df_ind_new.new_cases.plot(figsize=(20,10))
```

```
df_ind_new.plot(y=['new_cases','new_deaths'],figsize=(20,10))
```

```
df_ind_new['mortality_rate'] = df_ind_new['new_deaths']/df_ind_new['new_cases']*100
df_ind_new
```

```
df_ind_new.mortality_rate.plot(figsize=(15,5))
```

```
df_ind_new.plot(y=['new_cases','new_deaths','mortality_rate'],figsize=(15,5))
```

```
df_req = df_ind_new[['date','new_cases']]
```

```
df_req = df_req.rename(columns={'date':'ds','new_cases':'y'})
df_req
```

```
import fbprophet
model = fbprophet.Prophet(changepoint_prior_scale=0.7)
model.fit(df_req)
```

```
forecast = model.make_future_dataframe(periods=6,freq='M')
forecast = model.predict(forecast)
```

```
forecast
```

```
df_pre = forecast[['ds','yhat']]
```

```
df_pre
```

```
forecast[['ds','yhat','yhat_lower','yhat_upper']]
```

```
model.plot(forecast,xlabel='Date',ylabel='New_Cases');
```

```
from sklearn.metrics import r2_score
r2_score(df_req,df_pre);
```
